```sql
-- Customers (left table)
CREATE OR REPLACE TABLE customers (
  customer_id   INT,
  customer_name STRING
);

INSERT INTO customers VALUES
(1, 'Alice'),
(2, 'Bob'),
(3, 'Carol'),
(4, 'Dave'),
(5, 'Eve');

-- Orders (right table)
CREATE OR REPLACE TABLE orders (
  order_id    INT,
  customer_id INT,
  amount      INT
);

INSERT INTO orders VALUES
(101, 1, 500),
(102, 1, 300),
(103, 2, 200),
(104, 3, 100);
-- Note: customer_id 4 and 5 have no orders

-- LEFT SEMI: customers who have placed at least one order
SELECT
  c.customer_id,
  c.customer_name
FROM customers AS c
LEFT SEMI JOIN orders AS o
  ON c.customer_id = o.customer_id
ORDER BY c.customer_id;

-- LEFT ANTI: customers who have NEVER placed an order
SELECT
  c.customer_id,
  c.customer_name
FROM customers AS c
LEFT ANTI JOIN orders AS o
  ON c.customer_id = o.customer_id
ORDER BY c.customer_id;
```
